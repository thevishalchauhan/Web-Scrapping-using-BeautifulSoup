```python
from bs4 import BeautifulSoup
import requests

import pandas as pd
from pandas import Series,DataFrame
```

```python
url = 'http://www.ucop.edu/operating-budget/budgets-and-reports/legislative-reports/2013-14-legislative-session.html'
```

```python
result = requests.get(url)

c = result.content #all code of html comes here
```

```python
soup = BeautifulSoup(c,'html.parser')
#html parser is here
```

```python
summary = soup.find('div',{'class':'list-land','id':'content'})

tables = summary.find_all('table')
```

```python
data = []

rows = tables[0].findAll('tr')

for tr in rows:
    cols = tr.findAll('td')
    
    for td in cols:
        text = td.find(text = True)
        print(text, end=' ')
        data.append(text)
```

```python
data
```

```python
'''Now we'll use a for loop to go through the list and grab only the cells with a pdf file in them, 
   we'll also need to keep track of the index to set up the date of the report.'''

reports = []
date = []

index = 0

for item in data:
    if 'pdf' in item:
        date.append(data[index-1])
        
        reports.append(item.replace(u'\xa0',u''))
    index +=1
```

```python
date = Series(date)
reports = Series(reports)
```

```python
legislative_df = pd.concat([date,reports],axis=1)
```

```python
legislative_df.columns = ['Date','Report']
```
